```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt 
```

## Erosion

```python
# Grayscale image load
img = cv.imread("./images/sonar_image1.png", cv.IMREAD_GRAYSCALE)

# Compute Binary Map based on thresholding to get foreground
bin = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]

k_size = 4
# k_size kernel of ones 
k = np.ones((k_size,k_size), np.uint8)

# Invert Binary Map (Not needed)
inv = cv.bitwise_not(bin)

# Erosion Operator
out_erosion = cv.erode(bin, k, 1)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Eroded Image")
plt.imshow(out_erosion, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Bin Map Image")
plt.imshow(bin, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Inv Image")
plt.imshow(inv, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Dilation

```python
# Grayscale image load
# img = cv.imread("./images/sonar_image1.png", cv.IMREAD_GRAYSCALE)

# Compute Binary Map based on thresholding to get foreground
bin = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]

k_size = 4
# k_size x k_size  kernel of ones 
k = np.ones((k_size, k_size), np.uint8)

# Invert Binary Map (Not needed)
inv = cv.bitwise_not(bin)

# Dilation Operator
out_dilation = cv.dilate(bin, k, 1)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Dilated Image")
plt.imshow(out_dilation, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Bin Map Image")
plt.imshow(bin, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Inv Image")
plt.imshow(inv, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Opening

```python
# Compute Binary Map based on thresholding to get foreground
bin = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]

k_size = 4
# k_size x k_size  kernel of ones 
k = np.ones((k_size, k_size), np.uint8)


# Invert Binary Map (Not needed)
inv = cv.bitwise_not(bin)

# Opening Operator
out_opening = cv.morphologyEx(bin, cv.MORPH_OPEN, k)
```

```python
plt.figure(figsize=(10,8))

plt.subplot(2,2,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Opened Image")
plt.imshow(out_opening, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Bin Map Image")
plt.imshow(bin, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Inv Image")
plt.imshow(inv, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

## Summary

```python
# Apply ot directly to original image

# Grayscale image load
img = cv.imread("./images/sonar_image1.png", cv.IMREAD_GRAYSCALE)

# Compute Binary Map based on thresholding to get foreground
bin = cv.threshold(img, 0, 255, cv.THRESH_BINARY+cv.THRESH_OTSU)[1]

k_size = 3
# k_size kernel of ones 
k = np.ones((k_size,k_size), np.uint8)

# Erosion Operator
out_erosion = cv.erode(img, k, 1)
# Dilation Operator
out_dilation = cv.dilate(img, k, 1)
# Opening Operator
out_opening = cv.morphologyEx(img, cv.MORPH_OPEN, k)
# Closing Operator
out_closing = cv.morphologyEx(img, cv.MORPH_CLOSE, k)
# Gradient Operator
out_gradient = cv.morphologyEx(img, cv.MORPH_GRADIENT, k)
# Dilation Operator
out_tophat = cv.morphologyEx(img, cv.MORPH_TOPHAT, k)
# Black Hat Operator
out_blackhat = cv.morphologyEx(img, cv.MORPH_BLACKHAT, k)
```

```python
# Comparing these first methods: 

plt.figure(figsize=(10,8))

plt.subplot(2,4,1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2,4,2)
plt.title("Eroded Image")
plt.imshow(out_erosion, cmap='gray')
plt.axis('off')

plt.subplot(2,4,3)
plt.title("Dilated Image")
plt.imshow(out_dilation, cmap='gray')
plt.axis('off')

plt.subplot(2,4,4)
plt.title("Opened Image")
plt.imshow(out_opening, cmap='gray')
plt.axis('off')

plt.subplot(2,4,5)
plt.title("Closed Image")
plt.imshow(out_closing, cmap='gray')
plt.axis('off')

plt.subplot(2,4,6)
plt.title("Gradient Image")
plt.imshow(out_gradient, cmap='gray')
plt.axis('off')

plt.subplot(2,4,7)
plt.title("Top Hat Image")
plt.imshow(out_tophat, cmap='gray')
plt.axis('off')

plt.subplot(2,4,8)
plt.title("Black Hat Image")
plt.imshow(out_blackhat, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
