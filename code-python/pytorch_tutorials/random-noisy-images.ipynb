# Random Numbers to create Noise Image

## Random Tensors with PyTorch

```python
import torch
```

```python
# Random number with PyTorch 
# Random numers are foundamental in many ML workflow, from initializing weights and parameters
# and simulate real world variability, which is random in nature 

# It is possible to create an image noise with random numbers 

random_tensor = torch.rand(5) # 5 values between 0 and 1
print("Random tensor : ", random_tensor)
print("Random tensor dimension: ", random_tensor.ndim)
```

```python
# To modify the shape 
random_tensor = torch.rand(2, 5, 3) # 2 elemnts of 5 rows and 2 columns
print("Random tensor : ", random_tensor)
print("Random tensor dimension: ", random_tensor.ndim)
print("Random tensor shape: ", random_tensor.shape)
```

```python
# For reproducibility I can set a seed 
# Useful for testing to reproduce the same output 
torch.manual_seed(1) # write generic seed number 

# than create random number 
random_tensor = torch.rand(5) # 5 values between 0 and 1
print("Random tensor : ", random_tensor)

# Now every time the same random number will be created for reproducibility of result 
# See same values in between a team of developer or to have predictable random values 
```

```python
# To generate random integers 
random_tensor_int = torch.randint(low=-10, high=10, size=(2,2)) # high value is excluded in the possible values 

print("Random tensor : \n", random_tensor_int)
```

```python
# Extract random numbers from a normal distribution 

normal_tensor = torch.normal(mean=0.5, std=2, size=(5,))
print("Normal Distribution tensor : \n", normal_tensor)
```

## Noisy Images Creation

```python
# It is possible to use thid random extraction to generate random images

# Creating noise image using pytorch
import matplotlib.pyplot as plt

# h,w of image 256,256
# The image is a tensor in pytorch, w,h are the shape of the tensor 
uniform_noise = torch.rand((256,256))

plt.imshow(uniform_noise, cmap='grey', vmin=0, vmax=1)
plt.title("Uniform Noise Image")
plt.axis("off")
plt.show()

# In principe you can generate a video noise of it
```

```python
# To generate RGB noise 
uniform_noise_rgb = torch.rand((256,256,3))

plt.imshow(uniform_noise_rgb.numpy(), vmin=0, vmax=1) 
# You can even convert in numpy and it works
# It is better to work with numpy in image visualization since it is the standard
plt.title("Uniform Noise Image RGB")
plt.axis("off")
plt.show()

# In principe you can generate a video noise of it
```

```python
# take property of image 
print("Shape : ", uniform_noise_rgb.shape)
print("Data Type : ", uniform_noise_rgb.dtype)

print("Min val : ", torch.min(uniform_noise_rgb))
print("Max val : ", torch.max(uniform_noise_rgb))
```
