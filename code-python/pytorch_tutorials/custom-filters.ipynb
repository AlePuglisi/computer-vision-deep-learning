# Building custom filters 

```python
import numpy as np 
import cv2
from urllib.request import urlopen
from matplotlib import pyplot as plt
```

```python
def download_image(url):
    with urlopen(url) as resp: 
        image_data = np.asarray(bytearray(resp.read()), dtype="uint8")

    image = cv2.imdecode(image_data, cv2.IMREAD_COLOR)
    return image 

def display_image(image, cmap=None):
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image_rgb, cmap=cmap)
    plt.axis('off')
    plt.show()
```

```python
url = 'https://i.etsystatic.com/20636120/r/il/7724a4/2806797038/il_570xN.2806797038_7eln.jpg'  # Image URL
image = download_image(url)
display_image(image)
```

```python
# Canny edge 
edges = cv2.Canny(image,100,200)
display_image(edges)
```

```python
# VERTICAL EDGE DETECTOR
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([
    [-1,0,1],
    [-1,0,1],
    [-1,0,1]])

edge_detector = cv2.filter2D(image,-1,kernel)
display_image(edge_detector)
```

```python
# VERTICAL EDGE DETECTOR 2
kernel = np.float32([
    [1,0,-1],
    [1,0,-1],
    [1,0,-1]])

edge_detector = cv2.filter2D(image,-1,kernel)
display_image(edge_detector)
```

```python
# VERTICAL EDGE DETECTOR 3
kernel = np.float32([
    [-1,2,-1],
    [-1,2,-1],
    [-1,2,-1]])

edge_detector = cv2.filter2D(image,-1,kernel)
display_image(edge_detector)
```

```python
# HORIZONTAL EDGE DETECTOR 
kernel = np.float32([
    [-1,-1,-1],
    [2,2,2],
    [-1,-1,-1]])

edge_detector = cv2.filter2D(image,-1,kernel)
display_image(edge_detector)
```

```python
# HORIZONTAL EDGE DETECTOR 2
#kernel = np.ones((3,3),np.float32)/25
kernel = np.float32([
    [-1,-1,-1],
    [0,0,0],
    [1,1,1]])

edge_detector = cv2.filter2D(image,-1,kernel)
display_image(edge_detector)
```
