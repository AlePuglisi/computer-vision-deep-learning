# Working with file paths 

```python
import os
cwd = os.getcwd()
print(cwd)
```

```python
# create a new directory 
new_directory = "package"
os.mkdir(new_directory)

print(f'Directory {new_directory} is created')
```

```python
# List all the files/folders from a directory 
items = os.listdir('.') #. is the current working directory
print(items)
```

```python
# Joining paths with os
# path can come from different part, hardcoding is not a good practice (to avoid messing with /\ in windows or linux )
dir_name = "folder"
file_name = "file.txt"

# to create full relative path 
relative_path = os.path.join(dir_name, file_name)
print(relative_path)

# absolute path
absolute_path = os.path.join(os.getcwd(), dir_name, file_name)
print(absolute_path)
```

```python
path = 'example1.txt'
# Usefull to check if a file exists to have robustness
if os.path.exists(path):
    print(f"The file '{path}' exists")
else: 
    print(f"The file '{path}' does not exists")
```

```python
# Another way to check if a file exists: 
# you can check if a path is file or directory 
path = 'example.txt'

if os.path.isfile(path):
    print('file')
elif os.path.isdir(path):
    print('dir')
else:
    print('Not dir nor file')
```

```python
# get absolute path for any working dir
relative_path = 'example.txt'

# option 1
absolute_path = os.path.join(os.getcwd(), relative_path)
print(absolute_path)
# option 2
absolute_path = os.path.abspath(relative_path)
print(absolute_path)
```
