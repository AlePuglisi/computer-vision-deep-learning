# Python Operators

```python
## Many different opeartions for different data types is avaiable 

# Arithmetic 
a = 10
b = 3

add_result       = a + b
sub_result       = a - b
mult_result      = a * b
div_result       = a / b
floor_div_result = a // b
modulus_result   = a % b
exp_result       = a ** b

print(add_result)
print(sub_result)
print(mult_result)
print(div_result)
print(floor_div_result)
print(modulus_result)
print(exp_result)
```

```python
# Floor division
print(21 / 5)
print(21//5)
```

```python
## Comparison operators 
a = 10 
b = 10 

result = (a == b)
print(result)
```

```python
## Logical operators 
# and, not, or

x = True
y = True 

# and
result = x and y 
print(result)
```
