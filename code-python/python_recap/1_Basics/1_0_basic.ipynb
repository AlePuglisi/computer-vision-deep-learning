# Syntax and Semantics in Python

```python
# Basic Syntax Rules in Python 

## Case sensitity 

name = "Ale"
Name = "ale"

print(name)
print(Name)
# Python is case sensitive !
print(name != Name)
```

```python
## Indentation 
# Used to define blocks of code, 4 spaces (or tab) is used 
age = 24
if age > 30: 
    # Using indentation to identify if block code
    print(age)

print(age)
```

```python
## Line Continuation syntax 

# it is possible to continue code in multiple line using \
total = 1 + 1 + 1 + \
        1 \
        + 1

print(total)
```

```python
## Multiple statements on a single line 
x = 5; y = 10; z = x + y
# using semicolumn it is possible to do multiple statement in one line 
print(z)
```

```python
## Semantics in Python 

# variable assignment 
age = 24     # Kernel understand the variable type at runtime 
name = "Ale" # it is a string
```

```python
# Type Inference 
type(age)
```

```python
type(name)
```

```python
# Since typing occurs at runtime in python, dynamic typing is allowed...
variable = 10 
print(type(variable))
# dynamically change variable type (Inferencing)
variable = "Ale"
print(type(variable))
```
