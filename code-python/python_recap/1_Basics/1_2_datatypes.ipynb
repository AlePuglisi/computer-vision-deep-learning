# Python Data Types

```python
## Data Types tells the interpreter how the data is intended to use 
# It determine the amount of memory and values it can take
# Ensures data is stored efficently, manage memory, define operations allowed 

# Integers 
age = 24
type(age)
```

```python
# Floating point 
height = 1.72
print(height)
print(type(height))
```

```python
# String 
name = "Ale"
print(name)
print(type(name))
```

```python
# Boolean 
is_true = True 
print(is_true)
print(type(is_true))
```

```python
# Boolean 
is_true = bool()
print(is_true)
print(type(is_true))
# False by default
```

```python
# Boolean 
a = 10 
b = 10 

print(a == b)
print(type(a == b))
```

```python
# Common Errors in using data types 
#result = "Hello" + 5
# TypeError because concatenating different types 

# Type casting 
result = "Hello" + str(5)
print(result)
```
