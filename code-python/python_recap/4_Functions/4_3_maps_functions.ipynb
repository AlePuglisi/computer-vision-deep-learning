# map() function 

```python
# map() applies a given function to all the elemnts in an iterable, returning a map object (iterator)
# Useful for transforming data in a list comprehensively 

# example square function
def square(x):
    return x*x

square(6)
```

```python
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

# How to apply same operation to all elements in the list ? 
# using a loop in the square function is not the best option...
# It require multiple operations to be performed to obtain result. Instead we use map()

list(map(square, numbers)) # take each number and apply square function 
# map return an iterable map, so we convert it in a list 
```

```python
# Using lambda function with map 
numbers = [1, 2, 3, 4, 5, 6, 7, 8]

list(map(lambda x:x*x, numbers))
```

```python
# Can we map multiple iterables ? 
numbers1 = [1, 2, 3]
numbers2 = [4, 5, 6]

# I have two iterables, take each of both and add them between 
# I can do it with map and lambda 
added_number = list(map(lambda x,y: x + y, numbers1, numbers2))
# automatically extract vaue from more iterables
print(added_number)
```

```python
# map() to convert list of strings in integers 
str_numbers = ['1', '2', '3', '4', '5']
int_numbers =list(map(int,str_numbers)) # map function is int (in build function)

print(int_numbers)
```

```python
words = ["apple", "banana", "cherry"]
upper_word = list(map(str.upper, words))

# I can apply in built function such as str.upper
print(upper_word)
```

```python
# Example... I can apply map to list of dict 
def get_name(person):
    return person['name']

# This is good for one dict item ... If I have a list of dict items 
people = [
    {'name': "Ale", 'age': 24},
    {'name': "Sante", 'age': 21}
]

print(list(map(get_name, people)))
```
