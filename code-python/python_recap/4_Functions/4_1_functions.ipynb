```python
# Why functions ? 
num = 24
# check if even or odd
if num % 2 == 0: 
    print("the number is even")
else: 
    print("the numebr is odd")

# What if I need to reuse this piece of code multiple times ? 
# copy and paste same code ... 

# Reusing code with function help 
```

```python
# docstring are very important to organize code 
def even_or_odd(num):
    """This function finds even or odd"""
    print(f'The number {num} is', end = '')
    if num % 2 == 0: 
        print("even")
    else: 
        print("odd")
    
```

```python
even_or_odd(2)
even_or_odd(3)
even_or_odd(253)
```

```python
# decorators, lambda functions, etc... many possibility when defining functions

def greet(name):
    print(f'Hello {name}, Welcome ! ')

greet("Ale")
# greet() Raise Error if called without argument 
```

```python
# default parameters in functions 
# provide default values 
def greet(name = "Guest"):
    print(f'Hello {name}, Welcome ! ')

greet()
greet("customer")
```

```python
# Variable length arguments 
# 1. Positional arguments
# 2. Keywords arguments

# If I have many parameters in a function 
#def print_numbers(num1, num2, num3, num4, num5)

# 1. Positional arguments 
# instead positional arguments *args can be used 
def print_numbers(*args):
    for numbers in args: 
        print(numbers)

```

```python
print_numbers(1, 2, 3, 4, 5, 6, "Ale", "Wow")
# The input arguments value * reference all positional argumets received
# Useful when I have many arguments 
```

```python
# 2. Keywords arguments 
# This is given by **
# you use kwargs (keywords arguments)
def print_details(**kwargs):
    # all parameters in kwargs are in the form key:value pair !
    # So if it is in the form key:value you access it as dictionaries 
    for key, value in kwargs.items():
        print(f"{key}: {value}")

```

```python
print_details(name="Ale", age="24", country="Italy")
# All parameters are given in the form key=value to the function
# They are referenced by **kwargs
```

```python
# You can combine it !
def print_details(*args, **kwargs):
    for val in args: 
        print(f'Positional argument --> {val}')

    for key, value in kwargs.items():
        print(f"Keyword argument    --> {key}: {value}")
```

```python
# First all positional, then keywords args
print_details(1, 2, 3, 4, 5, "Ale", name="Ale", age="24", country="Italy")
```
