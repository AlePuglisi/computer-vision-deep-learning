# Lambda Functions

```python
# Small anonymous functions defined with lambda keyword
# (function without a name)

# they can have any number of argument but only one expression (only one logic)

#Syntax 
lambda arguments: expression
```

```python
def addition(a, b):
    return a + b

# maybe this small basic function with single logic/operation doesn't require a full function 
```

```python
addition_lambda = lambda a,b: a + b
# this is an anonymous function
type(addition_lambda)
```

```python
print(addition_lambda(5,6))
```

```python
# very simple way of defining a single usage simple expression
def even(num):
    if num % 2 == 0:
        return True

even_lamda = lambda num: num % 2 == 0

print(even_lamda(12))
```

```python
# Multiple parameters
def addition(x, y, z):
    return x + y + z
print(addition(1, 2, 3))

addition_lambda = lambda x,y,z: x + y + z
print(addition_lambda(1, 2, 3))

# No need of creating a whole function when a single expression is there
```

```python
# lambda functions have other features, requiring: Map and Filter functions (see soon)

# Example, map()
numbers = [1, 2, 3, 4, 5, 6]
# find square of each number 

def square(number):
    return number**2

# if I want to square over all the list I require an iteration (for loop) over the list ...

# As alternative, there is a better way with map() function
# map applies a function to all items in a list (collection)

# map require a function and an iterable where to apply the function
list(map(lambda x:x**2, numbers))
# create a map object ... 
# map uses a lazy loading .. to initialize we need to convert in list  
```
