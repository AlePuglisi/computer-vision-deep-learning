# Logging

```python
# Foundamental in ay python application to track events, errors and opeartions. 
# logging module of python allow emitting log messages etc
import logging
```

```python
# very useful for debugging, it allow many different log levels etc.. 
# AFter importing logging, the logging settings must be configured 

logging.basicConfig(level=logging.DEBUG)
# Debug level as minimum logging level visualized

# log messages with different severity levels 
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

```python
# Log levels indicate the severity of events.. Debug is used for diagnosis during implementation. 
# info is used to see that it is working as expected, warning when some problems is occured but not critical. 
# error and critical are serious problems 

# configure how logging messages are displayed 
logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```

```python
# to save logging details in a separate file 
import logging

logging.basicConfig(
    filename='app.log', 
    filemode='w',
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)

logging.debug("This is a debug message")
logging.info("This is an info message")
logging.warning("This is a warning message")
logging.error("This is an error message")
logging.critical("This is a critical message")
```
