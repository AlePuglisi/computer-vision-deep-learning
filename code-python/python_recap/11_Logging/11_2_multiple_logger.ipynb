# Logging with multiple loggers

```python
# create multiple loggers for different part of application 
import logging 

# I want to use this logger for Module1 of my app
logger1 = logging.getLogger("module1")
logger1.setLevel(logging.DEBUG)

# create a logger for module2
logger2 = logging.getLogger("module2")
logger2.setLevel(logging.WARNING)

logging.basicConfig(
    level=logging.DEBUG,
    format='%(asctime)s-%(name)s-%(levelname)s-%(message)s',
    datefmt='%Y-%m-%d %H:%M:%S'
)
```

```python
# log messages with different loggers 
logger1.debug("This is a debug message for module1")

logger2.debug("Try to launch debug message") # Not printed since minimum level is warning 
logger2.warning("This is a warning message for module1")

# in the log message it is evident written from which module the logging comes from (not just root)
```
