# Pandas DataFrame and Series 

```python
# Pandas is foundamental in data analysis
# pandas is used for many data manipulation operations 
# used also for cleaning, providing data structures series and data frames 
# Series: 1D array | DataFrame: 2D size-mutable heterogeneus tabular object with labeled axes (rows, columns)
# 
# numpy array can be converted into series, similarly 2D numpy array can be converted in data frame

import pandas as pd
```

```python
# Series 
# a series is a 1D array-like object that can hold any data type (similar to a column in a table)

data = [1, 2, 3, 4, 5]
series = pd.Series(data)
print(series)
print(type(series))
```

```python
# also you can create series from dict elements 
data = {'a': 1, 'b': 2, 'c': 3}

series_dict = pd.Series(data)
print(series_dict)
# the dict key become the indices, and the val become the element
```

```python
# I can use a different index based on req
data = [10, 20, 30]

# in excel files opening (csv) Pandas help in manipulating it
index = ['a', 'b', 'c']

series = pd.Series(data, index)
print(series)
```

```python
# DataFrame

# 2D mutable heterogeneus tabular data structured in row, columns... 
# similar to an excel sheet (multi dim data)
# I can read it easily and convert it

# it is possible to create a data frame from a dictionary of list 
data = {
    'Name':['Ale', 'John', 'Jake'], 
    'Age' :[24, 30, 45], 
    'City':['Catania', 'Florida', 'Messina']
}

df = pd.DataFrame(data)
print(df)
print(type(df))
```

```python
import numpy as np 
np.array(df)
```

```python
# I can convert in numpy array from pandas data frame 
# Create a data frame from a list of dictionaries 

data = [
    {'Name':'Ale', 'Age': 24, 'City': 'Catania'}, 
    {'Name':'Gio', 'Age': 26, 'City': 'Messina'}, 
    {'Name':'Bro', 'Age': 36, 'City': 'Palermo'},
    {'Name':'Ture', 'Age': 60, 'City': 'Siracusa'} 
]

df = pd.DataFrame(data)
print(df)
# Alternartive way!
```
