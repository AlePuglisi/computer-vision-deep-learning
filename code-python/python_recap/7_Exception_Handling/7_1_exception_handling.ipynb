```python
# Allow to handle errors and take corrective action without stopping program excecution 
# Avoid breaking program excecution, like wrong input etc...
# rise exception to solve error

# Exception are event that disrupt normal program flow. 
# occurs when Error during program excecution. 

a = 10 
a = b
# Name not define error!
# It is a NameError, which is a type of exception that can be handled.
```

```python
# Handling exception are good for code users of your applications 
# Exception try, except block 

try: 
    # try block is where something can occurs 
    a = b
except: 
    print("Variable has not been assigned")

# Exception is raised in except block 
```

```python
# We can display the specific exception
try: 
    # try block is where something can occurs 
    a = b
except NameError as ex: # I rely on the class responsible for that specific exception
    print(ex)
```

```python
# ZeroDivision is another type of exception you can handle 
try: 
    result = 1/0
except ZeroDivisionError as ex: 
    print(ex)
    print("Please eneter a valid denominator (not 0)")
```

```python
# Handle multiple exception
try: 
    result = 1/2
    a = b
except ZeroDivisionError as ex: 
    print(ex)
    print("Please eneter a valid denominator (not 0)")
# The last exception class generic should be in the last except (parent class handling all)
except Exception as ex1: 
    print(ex1)
    print("Main Exception encountered")
```

```python
try: 
    num = int(input("Enter a denominator number"))
    result = 10/num
except ValueError: 
    print("This is not a valid number")
except ZeroDivisionError:
    print("You can't enter denominator equal to 0")
except Exception as ex: # Handling any possible other error
    print(ex)
    print("Main Exception encountered")
```

```python
# try, except, else block 

try: 
    num = int(input("Enter a number"))
    result = 10/num
except ValueError: 
    print("This is not a valid number")
except ZeroDivisionError:
    print("You can't enter denominator equal to 0")
except Exception as ex: # Handling any possible other error
    print(ex)
# If No exception raise else block is excecuted
else:
    print("No error occurs !") 
    print(f"result = {result}")
```

```python
# try, except, else and finally 
try: 
    num = int(input("Enter a number"))
    result = 10/num
except ValueError: 
    print("This is not a valid number")
except ZeroDivisionError:
    print("You can't enter denominator equal to 0")
except Exception as ex: # Handling any possible other error
    print(ex)
# If No exception raise else block is excecuted
else:
    print("No error occurs !") 
    print(f"result = {result}")
finally: # Always excecuted at the end of the flow (even if exception is raised)
    print("Excecution complete.")

# A practical exemple is: connection to a database in the try block
# Any exception can occurs, than I need to close the conection. I need to be sure in the finally block that this is closed 
```

```python
# Example File Handling with Exception handling 
try: 
    file = open('example.txt', 'r')
    content = file.read()
    a = b
    print(content)
except FileNotFoundError: 
    print("The file does not exists")
except Exception as ex: 
    print(ex)
finally: 
    # Make sure the file is closed 
    if 'file' in locals() or not file.closed():
        file.close()
        print("File Closed")
```
