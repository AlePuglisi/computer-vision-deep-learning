# Generators

```python
# It is a subclass of iterators. 
# this allow to create iterators in a simple way, 
# using yield keyword to produce series of values lazily. Generating values on the fly, not storing in memory
# (again for efficient memory managment and looping)

def square(n):
    for i in range(n):
        return i**2

# In this way I'm returning just one value!
square(3)
```

```python
# I want to iterate through all elements
def square(n):
    for i in range(n):
        yield i**2

# yield create a local variables and saves it
square(3)
```

```python
# generators object create iterators in functions 
for i in square(3):
    print(i)
```

```python
# yield save the operations and return the value when calling it !
# it allow iteration and memory managment

# yield allow to access series of values lazily ! 
# No need of storing in memory ...
```

```python
a = square(3)
```

```python
# I can access the values as an iterator step by step 
# Generators create iterators thnks to yield keyword
# Whatever operation it return the value

def my_generator():
    yield 1
    yield 2
    yield 3

# I can return multiple yield value 
gen = my_generator()
gen
```

```python
next(gen)
```

```python
for val in gen: 
    print(val)

# It loop until yield values are available
# yield return next(gen) and you can loop through generator elements
```

```python
# Example 
# Reading large files

# useful in large files reading, allow to preocess one line at time without loading all in memory. 
# you load chanks of it, not all in memory !
# 
def read_large_file(file_path):
    with open(file_path, 'r') as file:
        for line in file: 
            yield line
# yield allow to get one line atime
```

```python
file_path = "large_file.txt"
for line in read_large_file(file_path):
    print(line.strip())
```
