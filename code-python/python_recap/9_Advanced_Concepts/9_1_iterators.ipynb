# Iterators

```python
# Advanced Python concept for efficient looping and memory managment
# access collection elements sequentially, without exposing the structure

my_list = [1, 2, 3, 4, 5, 6]
# to iterate though the list 
for i in my_list:
    print(i)

print(type(my_list))
```

```python
print(my_list)
my_list
```

```python
# How to create an iterators
iterator = iter(my_list) # becomes an iterator
print(type(iterator))
```

```python
# Efficient looping and Memory managment?
# in iteratir I cannot access all elements

# It is based on lazy loading technique...
# It is stored in memory, you can access values by iterating thorugh all its elements
print(next(iterator))
```

```python
# Lazy loading access one value per time
print(next(iterator))
```

```python
# I can access all elements with next() function
# Untile the last element, where if I call next I get an error

# Handling problem on access
try: 
    print(next(iterator))
except StopIteration:
    print("There are no other elements in the iterator")
```

```python
# The same can be done converting strings into iterators 
my_string = "Hello"
string_iterator = iter(my_string)

print(next(string_iterator))
```
