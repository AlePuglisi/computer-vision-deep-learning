```python
## Slicing lists
# interesting when accessing and modifing lists 
numbers = list(range(1,11))

print(numbers[2:5])
print(numbers[:5])
print(numbers[5:])
print(numbers[::2])  # move with step size of 2 
print(numbers[::-1]) # move with step size of -1 (opposite order)
print(numbers[::-2])
```

```python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
# Iterate over list (same as string)
for fruit in fruits:
    print(fruit)
```

```python
## Iterate with index (using enuumerate you get the index too)
for index,number in enumerate(numbers): 
    print(index, number)
```

```python
## List Comprehension 
# square all numbers in a specific range
lst = []
for x in range(10): 
    lst.append(x**2)

print(lst)
```

```python
# The same can be done in one line !
# (you can perform all this operation in a list definition)
lst = [x**2 for x in range(10)]
print(lst)

# What we want as output (formula x**2)
# What operation we want to perform (iteration)
```

```python
# In list comprehension, the syntax basics is the following 

# Basic
# [expression for item in iterable]
# expression is performed in all the elemets of the iterable

# + Conditional logic can be used
# [expression for item in iterable if condition]
# expression is performed in all the elements of iterable if condition is true

square = [num**2 for num in range(10)]
print(square)
```

```python
# select only even numbers 
lst = []
for i in range(10):
    if i % 2 == 0:
        lst.append(i)

print(lst)
```

```python
lst = [i for i in range(10) if i % 2 == 0]
print(lst)
```

```python
# You can nest list comprehension ! 

# + Nested list comprehension 
# [expression for item1 in iterable1 for item2 in iterable2]

lst1 = [1, 2, 3, 4]
lst2 = ['a', 'b', 'c', 'd']

pair = [ [i,j] for i in lst1 for j in lst2]
print(pair)

# Equivalent to 
pair.clear()
for i in lst1: 
    for j in lst2: 
        pair.append([i,j])
print(pair)
```

```python
# In build function can be used in list comprehension expression ! 
words = ["Hello", "World", "Python", "Comprehension"]
lengths = [len(word) for word in words]
print(lengths)
```

```python
# Else block can be used in list comprehension 
even_or_not = [True if i % 2 == 0 else False for i in range(10) ]
print(even_or_not)
```
