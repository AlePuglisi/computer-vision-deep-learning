# Touples

```python
# Immutable Nature !
# Ordered collection of items, immutable
# Similar to lists but immutability makes the differece ...

# Creating a touple 
empty_tuple = ()
print(empty_tuple)
print(type(empty_tuple))
```

```python
# Empty list 
lst = list()
print(type(lst))
# Empty tuple constructor
tpl = tuple()
print(type(tpl))
```

```python
# Fill a tuple 
numbers = tuple([1, 2, 3, 4, 5, 6])
print(numbers)
print(type(numbers))
```

```python
lst = list((1, 2, 3, 4, 5)) # Data casting
print(lst)
```

```python
mixed_tuple = (1, "Hello", 3.14, True)
print(mixed_tuple)
```

```python
# Accessing tuple elements 
print(numbers[0])
# by simple indexing !
```

```python
print(numbers[-1])
# same indexing as list is valid (also slicing)
```

```python
print(numbers[0:3])
```

```python
print(numbers[::-1])
```

```python
# same slicing and indexing rules of list ...
# tuple operations (concatenation, repetition, slicing)

# How to mix (concatenate)
concatenation_tuple = numbers + mixed_tuple
print(concatenation_tuple)
```

```python
# repetition
repeted_tuple = mixed_tuple * 3 
# get appended three times
print(repeted_tuple)
```

```python
numbers * 3 # also with numbers multiplication act as repetition
```

```python
# Immutable nature of tuples 
# meaning that elements cannot be changed ones assigned 
lst = [1, 2, 3, 4, 5]
print(lst)

lst[1] = "change"
print(lst)
# the list elements can be change ones assigned !
```

```python
numbers = tuple(lst)
numbers[1] = "change"
# ERROR, elements cannot be changed once assigned 
```

```python
# tuple methods 

# count occurences 
print(numbers.count(1))
```

```python
# at which index an element is present
# return first index of that value occurance 
print(numbers.index(4))
```

```python
print(numbers.index(10))
# Give error if not present
```

```python
# Packing and Unpacking tuple 
# useful when developing certain project
packed_tuple = 1, "Hello", 3.14
print(packed_tuple)
print(type(tuple))
# by default values with comma in between become tuple
```

```python
# Unpacking a tuple
# if we know a tuple has n elements
# I can create three variables and assign 
a, b, c = packed_tuple

print(a)
print(b)
print(c)
```

```python
# It is possible to unpack with *
numbers = (1, 2, 3, 4, 5, 6)
first, *middle, last = numbers
print(first)
print(middle)
print(last)

# I'm unpacking with less values than the lenght ...
# the *middle take all elements in between and store as a list  
```

```python
# Nested tuple
# tuple inside a tuple 

# Nested list 
lst = [[1, 2, 3, 4], [5, 6, 7, 8], [1, "Hello", 3.14, "d"]]
print(lst)

print(lst[0])
print(lst[0][:3]) # slicing in nested list 
```

```python
# tuple inside list 
lst = [[1, 2, 3, 4], [5, 6, 7, 8], (1, "Hello", 3.14, "d")]
print(lst[2][0:3])
```

```python
nested_tuple = ((1, 2, 3), ("a", "b", "c"), (True, False))
print(nested_tuple)
print(nested_tuple[0])
print(nested_tuple[1][2])
```

```python
# Iterating into nested tuple 
for tuple_index, sub_tuple in enumerate(nested_tuple): 
    print(f'sub tuple {tuple_index}')
    for item in sub_tuple: 
        print(item, end = "")
    print()
```
