# List Examples

```python
# To do list 
# Keep track of task you plan to do 

to_do_list = ["Buy Groceries", "Clean the house", "Pay bills"]

# Adding tasks 
to_do_list.append("Schedule meeting")
to_do_list.append("Go for a run")

# Removing tasks 
# after removing
to_do_list.remove("Clean the house")

# Check if a task is in the list 
if "Pay bills" in to_do_list:
    print("Hey don't forget to pay the utility bills")

print("To Do List remaining: ")
for task in to_do_list:
    print(f'-{task}')

# list methods helps you a lot to create such applications ! 
```

```python
# Organizing students grades 
grades = [85, 92, 75, 90, 88]

# Add new grade
grades.append(95)

# Calculate average
average_grade = sum(grades) / len(grades)
print(f"Average Grade: {average_grade}")

# Find highest and lowest (in build function)
highest_grade = max(grades)
lowest_grade = min(grades)
print(f"Highest Grade: {highest_grade}")
print(f"Lowest Grade: {lowest_grade}")
```
