# Conditional Statements 

```python
## if statement: 
# evaluate a condition and execute the block of code if condition is true 
age = 24

if age >= 18: 
    print("You are allowed to vote in the elections")
```

```python
## else statement 
# else execute block of code if the if condition is false
age = 16

if age >= 18:
    print("You are eligible for voting")
else: 
    print("You are a minor,you are NOT eligible for voting")

```

```python
## elif statement 
# allows to check multiple conditions, it stands for else if 
age = 20 

if age < 13: 
    print("You are a child")
elif age < 18: 
    print("You are a teenager")
else: 
    print("You are an adult !")
```

```python
## Nested conditional statements 
# combine if, else, elif inside blocks

# Check if number is even, odd, negative 
num = int(input("Enter a number ..."))

if num > 0: 
    print("The number is positive")
    if num % 2 == 0: 
        print("The number is even")
    else: 
        print("The number is odd")
elif num == 0: 
    print("The number is 0")
else: 
    print("The number is negative")

# take care ofthe indentation
```

```python
# Practical Example 
# determine if a year is leap year using nested conditional statement

year = int(input("Enter current year ..."))

if year % 4 == 0: 
    if year % 100 == 0: 
        if year % 400 == 0:
            print(year, " is a leap year")
        else:
            print(year, " is NOT a leap year")
    else: 
        print(year, " is a leap year")
else: 
    print(year, " is NOT a leap year") 
```

```python
## Exercise 
# Simple calculator program 
# take input and perform more operations... 

num1 = float(input("Enter first number ..."))
num2 = float(input("Enter second number ..."))
operation = input("Enter operation... +, -, /, *")

if operation == "+": 
    result = num1 + num2
elif operation == "-": 
    result = num1 - num2
elif operation == "/": 
    if num2 != 0:
        result = num1 / num2
    else: 
        print("Invalid Division by 0")
elif operation == "*": 
    result = num1 * num2
else: 
    print("Invalid operator")


print(num1, " ", operation, " ", num2, " = ", result)
```
