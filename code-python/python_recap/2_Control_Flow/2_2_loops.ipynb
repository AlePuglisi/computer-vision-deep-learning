# Loops 

```python
# Different loop to control the program flow ... 
# while, for, nested loop, break, continue, pass, etc ...

# for loop, used to iterate through a sequence of numbers 

# requires a sequence of numbers
range(5)

for i in range(5):
    print(i)
```

```python
for i in range(1, 6): 
    print(i)
```

```python
for i in range(1, 10, 2): 
    print(i)
```

```python
# Invert the order of the range with the step 
for i in range(10, 1, -1): 
    print(i)
```

```python
# Working with string 
name = "Ale Puglisi"
# A string is a collection of characters

for c in name: 
    print(c)
```

```python
## While loop 
# It is used to continue to execute a block until the condition is True

count = 0 

while count < 5: # Iterate until count < 5
    print(count)
    count += 1
```

```python
## Loop control statements 

# break 
# break statement exit the loop prematurely 
for i in range(10): 
    if i == 5: # condition to exit from the loop 
        break
    print(i)
```

```python
# continue 
# skips the current iteration and continues with the next 
for i in range(10): 
    if i % 2 == 0: 
        continue 
    print(i, " is odd")
```

```python
# pass
# null opeartion, it does nothing 
for i in range(5): 
    if i == 3: 
        #print("The number is ", i)
        pass # null statement... 
    print(i)
```

```python
# Nested loops 
# a loop inside another loop 

for i in range(3): 
    for j in range(2): 
        print(f"i: {i}, j:{j}")
```

```python
## Example 1

# sum of first N natual numbers using while loop and for loop 

N = int(input("Enter a natural number ..."))

result = 0

# While loop 
count = N
while count > 0:
    result += count
    count -= 1

# result = 0
# count = 1
# while count <= N:
#     result += count
#     count  += 1

print(f"sum of first {N} natural numbers = {result}")

# For loop 
result = 0
for i in range(N+1):
    result += i

print(f"sum of first {N} natural numbers = {result}")
```

```python
## Example 2
# display prime numbers between 1 and N 
N = int(input("Enter a natural number ..."))

for num in range(1, N+1): 
    if num > 1: 
        for i in range(2,num): 
            if num % i == 0:
                break
        else:
            print(num)
```
