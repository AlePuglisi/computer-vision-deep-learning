# Polymorphism

```python
# OOP concept that allow objects of different class of being threated as common superclass. 
# Allow to perform single action in different forms ! 
# Achieved through overriding and interfaces of methods 

# Method Overriding 
# Allow a child class to override a parent method, providing a specific implementation of a method
class Animal: 
    def speak(self):
        return "Sound of the Animal"
    
# Child class 1
class Dog(Animal):
    # Different implementation of a parent method
    def speak(self):
        return "Bau Bau !"

# Child class 2 
class Cat(Animal):
    def speak(self):
        return "Meow !"


# Function that demonstrate polymorphism
def animal_speak(animal): #object of class animal
    print(animal.speak())



dog = Dog()
print(dog.speak())

cat = Cat()
print(cat.speak())

animal_speak(dog) 
# the function adapt based on the instance...
# Polymorphism allow adapive behavior 

# Different behavior is guaranteed by the method overriding 
```

```python
# Polymorphism with functions and methods 
# Implemented using functions, to process in the same way objects of different types 
class Shape: 
    def area(slef):
        return "The area of the figure"

# Child class 1
class Reactangle(Shape):
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width * self.height

import math 
#Child class 2 
class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius
    
    def area(self):
        return self.radius**2 * math.pi


# Function that take a shape and compute area
# Demonstrate polymorphism 
def print_area(shape):
    print(f'The area is {shape.area()}')
```

```python
rectangle = Reactangle(10, 2)
circle = Circle(1)

print_area(rectangle)
print_area(circle)
```

```python
# Interfaces
# In python is Abstract base class !
# Plymorphism is help by abstract base class

# Abstract Base Class (ABC), define common methods for a group of related objects, 
# Enforce derived classes implement partcular methods with consistency

# Just define a blueprint withoud implementation

from abc import ABC, abstractmethod
# abstractmethod is a decorator ...

class Vehicle(ABC):
    # In abstract class, i define empty methods, just a skeleton
    # abstract method is inherited by all child
    @abstractmethod
    def start_engine(self):
        pass

# Child class 1
class Car(Vehicle):
    def start_engine(self):
        return "Car engine started"

# Child class 2 
class Motorcycle(Vehicle):
    def start_engine(self):
        return "Motorcycle engine started"
    

def start_vehicle(vehicle:Vehicle):
    print(vehicle.start_engine())

# Create objects of Car and Motorcycle 
car = Car()
motorcycle = Motorcycle()

start_vehicle(car)
start_vehicle(motorcycle)

```
