```python
# Instance Variables and Methods 
class Dog: 
    # constructor method (allow to define default construction)
    def __init__(self, name, age):
        # instance variable, available as part of the object
        # self is a keyword, responsible in identify istance variables !
        # outside of the class, identify the object itself. 
        self.name = name
        self.age = age
    
    # Instance Methods define function for the specific object 
    # I use self keyword so I have access to the attributes 
    def bark(self):
        print(f'Bau Bau ! ... {self.name} is barking !')
    

# create the object and exploit the attributes 
dog1 = Dog("Bobby", 1)
print(dog1)
print(dog1.name, dog1.age)
```

```python
dog2 = Dog("Lola", 3)
print(dog2)
print(dog2.name, dog2.age)

dog2.bark()
dog1.bark()
# Each object rely on same blueprint, but has its own value of the attributes 
```

```python
# Examample: Modeling a Bank account 

class BankAccount: 
    # Constructor
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance
    # Instance method
    def deposit(self, amount):
        self.balance += amount
        print(f'{amount} is deposited.\nNew balance is {self.balance}')
    
    def withdraw(self, amount):
        if amount >= self.balance: 
            print("Insufficient founds...")
        else: 
            self.balance -= amount
            print(f'{amount} withdrowed.\nNew balance is {self.balance}')
    
    def get_balance(self):

        return self.balance

account1 = BankAccount("John", 2000)
account1.deposit(100)
account1.withdraw(300)
print(f'Current balance: {account1.get_balance()}')
```
