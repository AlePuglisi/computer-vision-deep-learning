# Abstraction

```python
# Core concept that allow to hide complex implementation details. 
# Show only necessary features of objects. Helps reducing complexity and effort 

# All complex implementations remain hidden, the internal process is not important outside 
# to understand the high level working of an object. 
from abc import ABC, abstractmethod

# Define abstract class
class Vehicle(ABC):
    def drive(self):
        print("The vehicle is use for driving")

    # the abstract method is created such that the child class must implement
    # give you features that you exploit
    @abstractmethod
    def start_engine(self):
        pass


# The Car Child class inherit Vehicle 
# But you need to define the abstract method implementation
class Car(Vehicle):
    def start_engine(self):
        print("Car engine strated")


def operate_vehicle(vehicle:Vehicle):
    vehicle.start_engine()
    vehicle.drive()
    

car = Car()
operate_vehicle(car)
```
