# Magic Methods

```python
# Magic Methods: "dunder methods" (double underscore methods)
# special methods that strat and end with __, that define behavior for built in operations
# such as mathematical operations, comparison, print, etc

class Person: 
    pass

person = Person()
dir(person)
# dir show all magic methods, that we can override 

```

```python
print(person)
# the print is handled by __str__
# other methods can be overwritten
```

```python
class Person:
    # Basic init method
    def __init__(self, name, age):
        self.name = name
        self.age = age

person = Person("Ale", 24)
print(person)
# by default print give the object info...
```

```python
class Person:
    # Basic init method
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def __str__(self):
        return f'{self.name}, {self.age} years old'
    
    def __repr__(self):
        return f'Person(name={self.name}, age={self.age})'

# the behavior of object string return has been overwritten
person = Person("Ale", 24)
print(person)       # default string repr
print(repr(person)) # official string repr
```
