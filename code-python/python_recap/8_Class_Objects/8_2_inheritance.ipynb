# Inheritance in OOP

```python
# Inheritance allow to inherit attributes and methods from another class.
# Single and Multiple is allowed. 

# Parent class
class Car:
    def __init__(self, windows, doors, enginetype):
        self.windows = windows
        self.doors = doors
        self.enginetype = enginetype
    
    def drive(self):
        print(f"Brum Brum... from {self.enginetype} car")
    
    
    
```

```python
car1 = Car(4, 5, "petrol")
car1.drive()
```

```python
# A Tesla is a specific type of Car, I don't need to redefine it from scratch
# (Child class)
class Tesla(Car):
    def __init__(self, windows, doors, enginetype, is_self_driving):
        super().__init__(windows, doors, enginetype)
        self.is_self_driving = is_self_driving
    
    def drive(self): 
        if self.is_self_driving:
            print(f"Brum Brum... from {self.enginetype} autonomous car")
        else: 
            super().drive()
    
    def selfdriving(self):
        print(f"Tesla supports self_driving: {self.is_self_driving}")
```

```python
tesla1 = Tesla(4, 5, "Electric", True)
tesla1.selfdriving()
tesla1.drive()
```

```python
# It is possible to do multiple inheritance...
# When a class inherits from more than one base class 

# Base class 1 (parent 1)
class Animal: 
    def __init__(self, name):
        self.name = name

    def speak(self):
        print("Subclass must implement this method")

# Base class 2 (parent 2)
class Pet: 
    def __init__(self, owner):
        self.owner = owner

    def speak(self):
        print("Subclass must implement this method")

# Derived child 
class Dog(Animal, Pet):
    def __init__(self, name, owner):
        # When you have more parent class, you don't access it with super()
        Animal.__init__(self, name)
        Pet.__init__(self, owner)
    
    def speak(self):
        return f"{self.name} says Bau Bau !"
    

# Create an object and test
dog = Dog("Bobby", "John")
print(dog.speak())
print(f'Owner: {dog.owner}')
```
