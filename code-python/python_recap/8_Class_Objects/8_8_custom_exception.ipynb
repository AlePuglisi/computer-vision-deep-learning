# Custom Exception

```python
# define custom exception for error handling in your projects
# From a specifcic form, I want to rise and error custom exception if the user enter bad input


# This require to inherit from Exception class
class Error(Exception):
    pass 

# Generic Error custom exception
class dobException(Error):
    pass
```

```python
year = int(input("Enter the dob"))
import datetime
age = datetime.datetime.today().year - year

try:
    if age <= 30 and age >= 20:
        print("Valid age, you can apply")
    else:
        raise dobException
except dobException:
    print("Bad age")

```
