# Importing Modules

```python
# Modules and Packages 
# Provide lots of packages with many functions ready to use 
# Using this, you can use available functions, organize and reuse the code 

import math
# math is a package/module providing many in built math operation functions 
sq = math.sqrt(16)
print(sq)
```

```python
# I can import just a single function 
from math import sqrt, pi
# once imported like that: 
print(sqrt(64))
print(sqrt(pi))
print(pi)
```

```python
# How to install a package that you don't have
# numpy is used for tensor operations 
import numpy as np # np is an alias of numpy package, you can use it as np from here 
```

```python
# The venev doesn't have numpy installed ... 
# I install from requirements.txt !
# pip install -r requirements.txt

import numpy as np 
np.array([1, 2, 3, 4])
```

```python
from math import * 
# with * its importing all functions available in math 
# ready to use withoiut math.function notation 

print(pi)
print(sqrt(pi))
```

```python
# To create package available everywhere to order the code 
# 1. __init__.py is a special file 
# 2. Create my functions 
# Now I can easily import my custom modules 
from package.maths import addition
addition(2, 3)
```

```python
# similarly you can import it differently
from package import maths 
maths.addition(2, 3)
```

```python
# sub package import 
from package.subpackages import mult
mult.multiply(2, 3)
```
