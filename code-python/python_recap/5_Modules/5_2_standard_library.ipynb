```python
# Data Serialization 
# conevrting data into json and so on 
# json is important for API, when retrieving data from database etc. 
# return type and response often is in the form of json 
import json 
data = {'name': 'Ale', 'age': 24}

json_string = json.dumps(data)
print(json_string)
print(type(json_string))

# back to dict 
parsed_data = json.loads(json_string)
# when interacting with API receiving json, I can convert it back to dict for python usage 
print(parsed_data)
print(type(parsed_data))
```

```python
# csv 
# to work with csv file, pandas etc also allow this operations 
import csv 
# write a file (and create)
with open('example.csv', mode='w', newline='') as file: 
    writer = csv.writer(file)
    writer.writerow(['name', 'age'])
    writer.writerow(['Ale', 32])
# read a file
with open('example.csv', mode='r') as file: 
    reader = csv.reader(file)
    for row in reader: 
        print(row)
```

```python
# datetime (to access data values)
from datetime import datetime, timedelta

now = datetime.now()
print(now)

yesterday = now - timedelta(days=1)
print(yesterday)
```

```python
# time 
# to make program sleep for some time 
import time 
print(time.time())

# sleep for 2 sec
time.sleep(2)
print(time.time())
```

```python
# regular expression (re)
import re 
# this allow to match specific string pattern etc

pattern = r'\d+'
text = "There are 123 apples 456"
match = re.search(pattern, text)
# finish when mathces is encountered 

print(match.group())
# where digit are present... you extract that part of the text 
```
