# Image Filters

```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./images/filtering.jpg', cv.IMREAD_GRAYSCALE)

plt.imshow(image, cmap='gray')
```

```python
horizontal_kernel = np.array([[-1, -1, -1],
                               [0,  0,  0], 
                               [1,  1,  1]])

# to apply the horizontal filter, the convolution is computed automatically as
horizontal_filter_image = cv.filter2D(image, -1, horizontal_kernel)
plt.imshow(horizontal_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```

```python
# To get a better filtering, the intensity difference shoukld be changed to impact horizontal line effect

horizontal_kernel = np.array([[-1, -1, -1],
                               [5,  5,  5], 
                               [1,  1,  1]])

# to apply the horizontal filter, the convolution is computed automatically as
horizontal_filter_image = cv.filter2D(image, -1, horizontal_kernel)
plt.imshow(horizontal_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```

```python
# To get a better filtering, the intensity difference shoukld be changed to impact horizontal line effect
# By fine tuning the filter values is possible to increase the effect

horizontal_kernel = np.array([[-1, -1, -1],
                               [-10, -10, -10], 
                               [1,  1,  1]])

# to apply the horizontal filter, the convolution is computed automatically as
horizontal_filter_image = cv.filter2D(image, -1, horizontal_kernel)
plt.imshow(horizontal_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```

```python
# To get a better filtering, the intensity difference shoukld be changed to impact horizontal line effect
# By fine tuning the filter values is possible to increase the effect

x = -20

horizontal_kernel = np.array([[-1, -1, -1],
                               [x, x, x], 
                               [1,  1,  1]])

# to apply the horizontal filter, the convolution is computed automatically as
horizontal_filter_image = cv.filter2D(image, -1, horizontal_kernel)
plt.imshow(horizontal_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```

```python
# Vertical Kernel
x = -20

vertical_kernel = np.array([[-1, x, 1],
                            [-1, x, 1], 
                            [-1, x, 1]])

# to apply the horizontal filter, the convolution is computed automatically as
vertical_filter_image = cv.filter2D(image, -1, vertical_kernel)
plt.imshow(vertical_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```

```python
# Diagonal Kernel
x = -20

diagnoal_kernel = np.array([[x,-1, -1],
                            [1, x, -1], 
                            [1, 1, x]])

# to apply the horizontal filter, the convolution is computed automatically as
diagonal_filter_image = cv.filter2D(image, -1, diagnoal_kernel)
plt.imshow(diagonal_filter_image, cmap='gray')

#All horizontal lines are identified by white pixels...
# # But it is possible to increase the filter effect
```
