# Exploring Color Spaces

```python
import cv2 as cv
```

```python
bgr_image = cv.imread("./images/coral_reef.jpg")
```

```python
# Convert to GRAYSCALE
gray_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2GRAY)

cv.imwrite("./images/gray_image.jpg", gray_image)
```

```python
# Convert to HSV
# Not familiar for human "eye", the output of HSV and LAB is not familiar for human 
hsv_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2HSV)

cv.imwrite("./images/hsv_image.jpg", hsv_image)
# Not all human perceive the world... 
# HSV is not used for visualization, but for calculation, thresholding of color and other processing
```

```python
# Convert to LAB
# Not familiar for human "eye", the output of HSV and LAB is not familiar for human 
lab_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2LAB)

cv.imwrite("./images/lab_image.jpg", lab_image)
# Again not good for visualization but for calculation
```

```python
# BGR vs RGB
# Convert to RGB
rgb_image = cv.cvtColor(bgr_image, cv.COLOR_BGR2RGB)
# It is a basic swap of channels

cv.imwrite("./images/rgb_image.jpg", rgb_image)
# All red replaced by Blue !
```

```python
# Using the default color space BGR
bgr_image = cv.imread("./images/coral_reef.jpg")
cv.imwrite("./images/bgr_image.jpg", bgr_image)

# It uses the BGR but I see it properly as RGB !
# By saving it from BGR I see it as RGB ... Why ? 

# OpenCV automatically change the channels from BGR and RGB, we are overriding the input expected as RGB, 
# A double swapping occurs by internally changing channel order in the unexpected way
```
