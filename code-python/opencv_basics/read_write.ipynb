```python
import cv2 
```

```python
# Read the Image

image = cv2.imread("./images/coral_reef.jpg")
```

```python
# check if image is loaded sucessfully 

if image is None: 
    print("Error : Could not load the image")
else: 
    print("Image loaded sucesfully")
```

```python
# Display the image in the notebook 

cv2.imshow("Image Window", image) # Title of diplay window to show, and argument of image
cv2.waitKey(0)                    # pauses program until a key is pressed (wait open until input)
cv2.destroyAllWindows()           # destroy all when key is pressed 
```

```python
# Write an Image 

image_new = image + 30 # add 30 to all pixels
cv2.imwrite("./images/output.jpg", image_new)
```

```python
# Display the New image in the notebook 

cv2.imshow("Image Window", image_new) # Title of diplay window to show, and argument of image
cv2.waitKey(0)                    # pauses program until a key is pressed (wait open until input)
cv2.destroyAllWindows()           # destroy all when key is pressed 
```
