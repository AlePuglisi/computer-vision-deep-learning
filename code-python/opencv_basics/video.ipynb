# Working with video files

```python
import cv2
```

```python
# Load the Video 
# The video capture object is not the video itself. 
# This object help us to retrieve the frames from the video one by one
video = cv2.VideoCapture("./videos/coral_reef.mp4")

```

```python
# Display the Video 
while True: 
    ret, frame = video.read() # read method on video capture
    # ret is true when sucessfully capture the frame, otherwise if no frame left I get false
    # frame instead get the image
    if not ret: 
        break

    # Display the Image
    cv2.imshow("Video Frame", frame)

    # How to close the while loop

    # we wait 1 ms between frames as waitKey(1), so we show next frame only every 1ms
    # by default waitKey(1) return -1 if we don't press anything 

    # when something is pressed on the keyboard, it return ASCII of that value, as component of that 
    # That bitwise & 0xFF trim to 8 digits, after pressing something we want to stop
    # we set q as stopping key from keyboard

    if cv2.waitKey(30) & 0xFF == ord('q'):
        break


# release video capture and destroy all windows
video.release()
cv2.destroyAllWindows()
# Be sure of releasing the video capture and destroy the window, or you brake the code...
```

```python
video = cv2.VideoCapture("./videos/coral_reef.mp4")
# Write the video to local storage

# Here the 4cc, codec algorithms cames into play
fourcc = cv2.VideoWriter_fourcc(*"mp4v")

width = int(video.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(video.get(cv2.CAP_PROP_FRAME_HEIGHT))

out = cv2.VideoWriter("./videos/output.mp4", fourcc, 30, (width, height))

while True: 
    ret, frame = video.read() # read method on video capture

    if not ret: 
        break

    out.write(frame)

out.release()
video.release()
cv2.destroyAllWindows()
```
