# Flipping, Rotating, Cropping

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
image = cv.imread("./images/coral_reef.jpg")
image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)

print(image)
```

```python
# Flipping Image
# flip axis = 1 : Horizontal flip (Around vertical axis)
# flip axis = 0 : Vertical flip  (Around Horizontal axis)
# flip axis = -1 : Both axis flip (H (Around Vertical) + V (around Horizontal))
flip_horizontal = cv.flip(image, 1)
flip_vertical   = cv.flip(image, 0)
flip_both       = cv.flip(image, -1)

titles = ["Original Image", "Horizontal Flip", "Vertical Flip", "Both Axis Flip"]
images = [image_rgb,
          cv.cvtColor(flip_horizontal, cv.COLOR_BGR2RGB),
          cv.cvtColor(flip_vertical, cv.COLOR_BGR2RGB),
          cv.cvtColor(flip_both, cv.COLOR_BGR2RGB)]

plt.figure(figsize=(12,8))
for i in range(len(titles)):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Rotations 

# Basic rotation
rotate_90cw_image = cv.rotate(image_rgb, cv.ROTATE_90_CLOCKWISE)
rotate_90ccw_image = cv.rotate(image_rgb, cv.ROTATE_90_COUNTERCLOCKWISE)
rotate_180_image = cv.rotate(image_rgb, cv.ROTATE_180)

titles = ["Original Image", "Rotate 90 CW", "Rotate 90 CCW", "Rotate 180"]
images = [image_rgb,
          rotate_90cw_image,
          rotate_90ccw_image,
          rotate_180_image
        ]

plt.figure(figsize=(12,8))
for i in range(len(titles)):
    plt.subplot(2, 2, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Cropping 

print(image_rgb.shape)
```

```python
# once we know the shape, select the two coordinates of the cropping box
x_start, y_start = 80, 300
x_end  , y_end   = 150, 380

# slicing by selecting coordinates with y slice, x slice
# in opencv coordinates start from top left with x axis horizontal towards the right, y axis vertical to bottom
cropped_image = image_rgb[y_start:y_end, x_start:x_end]

plt.figure(figsize=(8,4))

plt.subplot(1, 2, 1)
plt.imshow(image_rgb)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cropped_image)
plt.title("Cropped Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```
