# Drawing Lines and Shapes

```python
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create a blank canvas where drawing shapes and lines 
height, width = 500, 700
blue = (255, 127, 0)       # blue background of the canva

canvas = np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
# without converting to RGB this will be in BGR format as orange !
```

```python
# Draw a Line
canvas_rgb = cv.cvtColor(canvas, cv.COLOR_BGR2RGB)
cv.line(canvas_rgb, (50, 50), (650, 450), 
                     color=(255, 255, 255), 
                     thickness=5)

plt.imshow(canvas_rgb)

# NOTE: everytime you rerun this cell a new line is added since the image itself is overwritten !
```

```python
# Draw a Rectangle
cv.rectangle(canvas, (100, 100),(300, 300), 
             color=(0, 255, 255),
             thickness=3) # thickness = -1 will fill the whole shape 

# with tickness -1 it fill the shape 
cv.rectangle(canvas, (150, 150),(250, 250), 
             color=(0, 0, 255),
             thickness=-1) 

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# Draw a Circle 
cv.circle(canvas, (500, 150), 50,
          color=(0,255,0),
          thickness=3)
# again if tickness = -1 it is filled circle
cv.circle(canvas, (500, 150), 20,
          color=(0,255,0),
          thickness=-1)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# Draw an Ellipse 
cv.ellipse(canvas, (350, 350), 
           axes=(100, 50),
           angle=45, startAngle=0, endAngle=360,
           color=(0,255,127),
           thickness=3)

cv.ellipse(canvas, (350, 350), 
           axes=(50, 25),
           angle=45, startAngle=180, endAngle=360,
           color=(0,255,127),
           thickness=-1)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# To draw more complex shapes such as polygons
# a Polygon is a set of multiple pints Xn,Yn with n > 2
# minimum amount of points to make a line is 3
P = np.array([[150, 400], [300, 400], [225, 300]], np.int32)
P_reshape = P.reshape((-1, 1, 2))

print("P: \n ", P)
print("\nP reshaped: \n", P_reshape)
# the reshaped version is the one expected by openCV to create a polyhon
# reshape(-1,1,2) is used to 
# single row 1 put to two values 2 in auto mode (-1)
# reshape in 1 row and 2 column in each shape (automatically)

cv.polylines(canvas, [P_reshape], 
             isClosed=True, # specify if closing the loop or not with the specified points (connecting first and last)
             color=(0,255,0),
             thickness=3
             )

cv.polylines(canvas, [P_reshape], 
             isClosed=False, # specify if closing the loop or not with the specified points (connecting first and last)
             color=(0,0,255),
             thickness=3
             )
plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
P = np.array([[150, 400], [300, 400], [225, 300]], np.int32)
P_reshape = P.reshape((-1, 1, 2))

print("P (shape = ", P.shape, ") : \n ", P)
print("\nP reshaped (shape = ", P_reshape.shape, ") :\n", P_reshape)
```

# Adding Text

```python
# Again lets create a blank canva 
height, width = 500, 700
blue = (255, 127, 0) # Sky blue color

canvas = np.full((height, width, 3), blue, dtype=np.uint8)

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# Add text 
text = "Hello Blue"
font = cv.FONT_HERSHEY_COMPLEX
# The orh is the coordinate for the bottom left corner of the text 
# it is the origin of the text
org  = (50, 100)
font_scale = 1.5 # by default it is 14, the scale is the product factor 
# with scale 1.5 it become 14*1.5 = 21
color = (255, 255, 255)
thickness = 2

cv.putText(canvas,text,org, font,
           font_scale,color,thickness,
           cv.LINE_AA) #Line_AA define the type of line used for rendering the text in the image 


plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# Fonts Types
# Again lets create a blank canva 
height, width = 500, 700
blue = (255, 127, 0) # Sky blue color

canvas = np.full((height, width, 3), blue, dtype=np.uint8)

# experiment with different fonts
fonts = [
    cv.FONT_HERSHEY_SIMPLEX,
    cv.FONT_HERSHEY_COMPLEX,
    cv.FONT_HERSHEY_SCRIPT_SIMPLEX,
    cv.FONT_HERSHEY_SCRIPT_COMPLEX,
    cv.FONT_HERSHEY_COMPLEX_SMALL
]

y_offset = 50

for i, font in enumerate(fonts):
    text = f'Font {i+1}'
    cv.putText(canvas, text, (50, y_offset), font, 1, (0,255,0), 2, cv.LINE_AA)
    y_offset += 50

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
```

```python
# Font Size and Thikness

# Again lets create a blank canva 
height, width = 500, 700
blue = (255, 127, 0) # Sky blue color

canvas = np.full((height, width, 3), blue, dtype=np.uint8)
styles = [
    (1, 2),
    (2, 2),
    (1, 4),
    (2, 4),
]

y_offset = 50

for font_scale, thickness in styles:
    text = f'Font {i+1}'
    cv.putText(canvas, text, (50, y_offset), cv.FONT_HERSHEY_SIMPLEX, font_scale, (0,255,0), thickness, cv.LINE_AA)
    y_offset += 50

plt.imshow(cv.cvtColor(canvas, cv.COLOR_BGR2RGB))
    
```
