# Blur Filters: Median, Gaussian, Average

```python
import cv2 as cv
import numpy as np 
import matplotlib.pyplot as plt
```

```python
image = cv.imread('./images/coral_reef2.jpg')

plt.imshow(cv.cvtColor(image, cv.COLOR_BGR2RGB))

image_rgb = cv.cvtColor(image, cv.COLOR_BGR2RGB)
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
```

```python
# Apply Blur on the image

# Average Blur
kernel_s = 10
average_blur_img = cv.blur(image_gray, (kernel_s, kernel_s))

plt.imshow(average_blur_img, cmap='gray')
```

```python
# Gaussian Blur
kernel_s = 15 # Kernel size must be an odd number 
gaussian_blur_img = cv.GaussianBlur(image_gray, (kernel_s, kernel_s), 
                                   sigmaX=5, sigmaY=5)

plt.imshow(gaussian_blur_img, cmap='gray')
```

```python
# Gaussian Blur
kernel_s = 11 # Kernel size must be an odd number 
median_blur_img = cv.medianBlur(image_gray, kernel_s)

plt.imshow(median_blur_img, cmap='gray')
```

```python
plt.figure(figsize=(14,12))

plt.subplot(2,2,1)
plt.title("Original Grayscale")
plt.imshow(image_gray, cmap='gray')
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Average Blur")
plt.imshow(average_blur_img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Gaussian Blur")
plt.imshow(gaussian_blur_img, cmap='gray')
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Median Blur")
plt.imshow(median_blur_img, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```
